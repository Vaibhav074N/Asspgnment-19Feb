Q3.What is App routing in Flask? Why do we use app routes?

Ans:

In Flask, an app route is a way to map a URL to a particular Python function. When a user visits a specific URL, Flask uses the app route to determine which function to call to generate the appropriate response.
App routing in Flask is done using the @app.route() decorator. 

We use app routes in Flask to provide a logical structure to our web application. By mapping specific URLs to specific functions, we can create a set of routes that provide the functionality required by the application. This helps to keep our code organized and easy to maintain.

Q5.What function is used in Flask for URL Building? 

Ans:

The url_for() function in Flask is used for URL building. It generates a URL for a given endpoint with the given arguments. The function takes the name of the endpoint as its first argument and any keyword arguments that represent variable parts of the URL

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def i():
    return 'This is the me.'

@app.route('/<instaid>')
def profile(instaid):
    return f'This is the profile page for {instaid}'

@app.route('/<int:post_id>')
def post(post_id):
    return f'This is the page for post {post_id}'

with app.test_request_context():
    print(url_for('i'))  
    print(url_for('profile', instaid='vaibhav'))  
    print(url_for('post', post_id=101))  
```
